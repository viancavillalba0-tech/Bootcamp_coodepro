```python
class Personaje:
    def __init__(self, nombre, super_poder):
        self.nombre = nombre
        self.super_poder = super_poder 

    def presentarse(self):
        print(f"Hola, soy {self.nombre} y mi súper poder es {self.super_poder}.")

class SuperPenguin(Personaje):
    def volar(self):
        print(f"{self.nombre} está volando a una altura de 10 metros.")

class Reven(Personaje):
    def usar_telequinesis(self):
        print(f"{self.nombre} está moviendo el vaso con la mente.")


heroe = SuperPenguin("Super Penguin", "vuelo")
heroe.presentarse()


villano = Reven("Reven", "telequinesis")
villano.presentarse()
```

```python
class Animal:
    especie = "mamifero"
    
    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color
    def son (self, sonido):
        self.sonido = sonido 
    def mue (self, mueve):
        self.mueve = mueve

animal1 = Animal("vaca", "blanca")
animal1.son("muuu")
animal1.mue("pasos")

print(f"soy una {animal1.nombre} digo {animal1.sonido} y doy {animal1.mueve}")

```
