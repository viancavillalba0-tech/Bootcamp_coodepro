```python
precio_monto = int(input("ingresa el precio"))
impuesto = 10
def calcular_el_total (precio):
    total = precio + precio*1/impuesto 
    print (f"el precio total es {total} gs")

calcular_el_total(precio_monto)
```

```python
cantidad_de_flor = int(input("ingresa la cantidad de flores que quieres en tu ramo"))
tipo_de_flor = input("ingresa el tipo de flor que quieres")
color_usuario = input("ingresar el color que quieres que sea tu flor")
def preparar_pedido (color, flor, cantidad):
    print(f"preparando tu ramo de {cantidad} de {flor} de color {color}")
    print("🌸🌺"*cantidad)
    
preparar_pedido(color_usuario, tipo_de_flor, cantidad_de_flor)
```

```python
cantidad_total_de_caramelos = int(input("cantidad de caramelos"))
cantidad_de_niños = int(input("cantidad de niños"))
#divicion 
cuanto_recibe_cada_niño = cantidad_total_de_caramelos // cantidad_de_niños
print ("cada niño recibe:", cuanto_recibe_cada_niño)
#modulo
cuanto_sobra = cantidad_total_de_caramelos % cantidad_de_niños
print ("sobra", cuanto_sobra)
```
