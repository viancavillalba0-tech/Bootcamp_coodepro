```python
tarjeta_personal ={}
def crear ():
    clave =input("ingresar el nombre de la clave")
    dato =input("ingresa el dato")
    tarjeta_personal[clave]= dato
    print(tarjeta_personal)
    
def modificar ():
    clave =input("ingresar el nombre de la clave a modificar")
    dato =input("ingresa el dato a modificar")
    tarjeta_personal[clave]= dato
    print(tarjeta_personal)

def eliminar ():
    
    clave =input("ingresa el dato a eliminar")
    del tarjeta_personal[clave]
    print(tarjeta_personal)
   


while True:
    opcion = int(input("Ingresar una opcion: 1. Crear 2. Modificar 3. eliminar 4. salir "))
    
    if opcion == 1:
        crear ()
        
    elif opcion == 2:
        modificar ()
       
    elif opcion == 3:
        eliminar ()
    else:
        break
    
    
```
